# Part 1: Data Loading and Basic Exploration 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/metadata.csv', nrows=10000)
```

```python
# 1. Basic Info
df.info()
```

```python
print("Shape:", df.shape)
df.head()
```

```python
# 2. Missing Values
missing = df.isnull().sum().sort_values(ascending=False)
print("Missing Values:\n", missing)
```

```python
# 3. Basic Statistics
print("Basic Statistics:\n", df.describe(include='all'))
```

# Part 2: Data Cleaning and Preparation

```python
# 1. Dropping columns with 100% missing values
cols_to_drop = missing[missing == len(df)].index
df_cleaned = df.drop(columns=cols_to_drop)
```

```python
df_cleaned.info()
print("Cleaned Shape:", df_cleaned.shape)
df_cleaned.head()
```

```python
# 2. Handle key columns
df_cleaned['abstract'] = df_cleaned['abstract'].fillna('No Abstract')
df_cleaned['authors'] = df_cleaned['authors'].fillna('Unknown')
df_cleaned['journal'] = df_cleaned['journal'].fillna('Unknown')
```

```python
# 3. Drop columns to reduce noise
cols_to_drop = ['sha', 'pdf_json_files', 'pmc_json_files']
df_cleaned = df_cleaned.drop(columns=cols_to_drop)
df_cleaned.info()
```

```python
# Prepare for further analysis
# Convert 'publish_time' to datetime
df_cleaned['publish_time'] = pd.to_datetime(df_cleaned['publish_time'], errors='coerce')
df_cleaned['year'] = df_cleaned['publish_time'].dt.year
```

```python
# Add abstract word count column
df_cleaned['abstract_word_count'] = df_cleaned['abstract'].apply(lambda x: len(x.split()))
```

```python
# Save cleaned csv
df_cleaned.to_csv('../data/metadata_cleaned.csv', index=False)
```

# Part 3: Data Analysis and Visualization

```python
# Count papers by publication year
year_counts = df_cleaned['year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.barplot(x=year_counts.index, y=year_counts.values, color='skyblue')
plt.title('Publications by Year')
plt.xlabel('Year')
plt.ylabel('Number of Publications')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Identify top journals
top_journals = df_cleaned['journal'].value_counts().head(10)
plt.figure(figsize=(8, 5))
sns.barplot(x=top_journals.index, y=top_journals.values, color='lightgreen')
plt.title('Top 10 Journals by Number of Publications')
plt.xlabel('Journal')
plt.ylabel('Number of Publications')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Most frequent words in titles
from collections import Counter
import re

def word_freq(texts):
    words = []
    for text in texts.dropna():
        words.extend(re.findall(r'\b\w+\b', str(text).lower()))
    return Counter(words).most_common(20)

top_words = word_freq(df_cleaned['title'])
print("Top 20 Words in Titles:\n", top_words)
```

```python
# 4. Word cloud of paper titles
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Generate a word cloud of all paper titles
titles_text = " ".join(df_cleaned['title'].dropna().tolist())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(titles_text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Paper Titles")
plt.show()
```

```python
# 5. Distribution of paper counts by source
source_counts = df_cleaned['source_x'].value_counts()
plt.figure(figsize=(8,5))
sns.barplot(x=source_counts.index, y=source_counts.values, color='salmon')
plt.title('Distribution of Paper Counts by Source')
plt.xlabel('Source')
plt.ylabel('Number of Papers')
plt.xticks(rotation=45)
plt.show()
```
